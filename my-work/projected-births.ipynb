Read from projectedbirths-cso.csv and plot the data

```python
# Projected Birth Rates in Ireland
import pandas as pd
import matplotlib.pyplot as plt

# Load the projected births data
births = pd.read_csv('data/projectedbirths-cso.csv')

# Display the first few rows
display(births.head())

# Plot the projected annual births
plt.figure(figsize=(10,6))
plt.plot(births['Year'], births['VALUE'], marker='o')
plt.title('Projected Annual Births in Ireland (CSO, Method M1)')
plt.xlabel('Year')
plt.ylabel('Number of Births')
plt.grid(True)
plt.show()
```
