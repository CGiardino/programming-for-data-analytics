# Assignment 3: Pie Charts
This notebook will download the data of 1000 people in csv format, extract the domains from their email addresses, count the occurrences of each domain, and plot a pie chart to visualize the distribution of email domains.

## Step 1:
Importing data from a web URL and extracting email domains

```python
# Import pandas to handle data
import pandas as pd
# Import matplotlib for plotting
import matplotlib.pyplot as plt
# Import requests to download data
import requests
# Import io to handle in-memory file
import io

# Download the csv data
# https://www.w3schools.com/python/ref_requests_get.asp
url = 'https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download'
people_data = requests.get(url)

# Read the CSV data into a DataFrame
# https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html
people_df = pd.read_csv(io.StringIO(people_data.text))
```

## Step 2:
Extracting email domains and counting occurrences

```python
# Extract the domains from email addresses by splitting at '@' and taking the second part
# https://pandas.pydata.org/docs/reference/api/pandas.Series.apply.html
# https://www.w3schools.com/python/ref_string_split.asp
people_df['Domain'] = people_df['Email'].apply(lambda x: x.split('@')[1])

# Count occurrences of each domain
# https://pandas.pydata.org/docs/reference/api/pandas.Series.value_counts.html
domain_counts = people_df['Domain'].value_counts()
```

## Step 3:
Plotting the pie chart with emailSt domain distribution

```python
# Create labels with counts for the pie chart by combining domain names and their counts
# https://www.w3schools.com/python/python_for_loops.asp
# https://www.w3schools.com/python/ref_func_zip.asp
labels = [f"{domain} ({count})" for domain, count in zip(domain_counts.index, domain_counts.values)]

# Setting up the pie chart with a larger figure size
# https://www.w3schools.com/python/matplotlib_pie_charts.asp
plt.figure(figsize=(10, 8))
# Setting up the pie chart with domain usage percentages and larger font size
plt.pie(domain_counts, autopct='%1.2f%%', textprops={'fontsize': 15})
# Adding title
plt.title('Email Domain Distribution')
# Ensuring the pie chart is a circle
plt.axis('equal') 
# Adding legend with labels with domain names and counts
plt.legend(labels=labels, title="Domains and Counts", loc="upper right", bbox_to_anchor=(1.2, 1))
# Display the pie chart
plt.show()
```

## Description of the pie chart
The pie chart illustrates the distribution of email domains among a sample of 1000 people.
Each slice represents a different email domain, with the percentage of users for that domain.

The distribution is very even, with example.org being slightly more common.
